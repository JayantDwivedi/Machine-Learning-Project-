```python
import pandas as pd
```

```python
data = pd.read_csv('Uber Review Data.csv',encoding="ISO-8859-1")
```

```python
data
```

```python
def sentiment(df):
    if (df['Stars'] > 3):
        return "best"
    elif (df['Stars'] < 3):
        return "worst"
    if(df['Stars'] == 3):
        return "average"
```

```python
data['Sentiment'] = data.apply(sentiment,axis=1)
```
